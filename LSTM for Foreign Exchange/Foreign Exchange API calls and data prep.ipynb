```python
#Open the kaggle data set
historical = pd.read_csv('Foreign_Exchange_Rates.csv')
historical.head()
```

```python
#remove irrelevant columns
historical = historical[['Time Serie', 'AUSTRALIA - AUSTRALIAN DOLLAR/US$', 'EURO AREA - EURO/US$', 'CANADA - CANADIAN DOLLAR/US$', 'UNITED KINGDOM - UNITED KINGDOM POUND/US$', 'SWITZERLAND - FRANC/US$']]
#rename columns
historical = historical.rename(columns={"Time Serie": "Date", "AUSTRALIA - AUSTRALIAN DOLLAR/US$": "AUD", 'EURO AREA - EURO/US$': 'EUR', 'UNITED KINGDOM - UNITED KINGDOM POUND/US$': 'GBP', 'CANADA - CANADIAN DOLLAR/US$': 'CAD', 'SWITZERLAND - FRANC/US$': 'CHF'})
#check
historical.head()
```

```python
#drop any null values
import numpy as np
historical = historical.replace('ND', np.nan)
historical = historical.dropna().reset_index(drop=True)
historical.isna().sum()
#check
historical.tail()
```

```python
#rest index for recent years df
recent_years = recent_years.reset_index()
#check
recent_years.head()
```

```python
#rename the index column
recent_years = recent_years.rename(columns={'index': 'Date'})
#check
recent_years.head()
```

```python
#concatenate the historical and recent data sets
all_years = pd.concat([historical, recent_years])
#check
all_years.head()
```

```python
all_years.tail()
```

```python
#set date as index again to help resize df to 20 years of data
all_years = all_years.set_index('Date')
#check
all_years.head()
```

```python
#Narrow down to 20 years worth of data: 2003 through 2022
final_df = all_years.loc['2003-01-01': '2023-01-01']
#check
final_df.head()
```

```python
#check for all dates
final_df.tail()
```

```python
#check number of records
len(final_df)
```

```python
#enter final df into a csv
final_df.to_csv('Forex_20_years.csv')
```
